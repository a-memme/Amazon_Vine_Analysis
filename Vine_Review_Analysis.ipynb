```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df
```

```python
# New dataframe where total_votes >= 20
votes_df = vine_df[vine_df["total_votes"]>= 20]
votes_df.head(30)
```

```python
votes_df.dtypes
```

```python
# New dataframe where number of helpful_votes / total_votes is greater or equal to 50%
new_votes_df = votes_df[votes_df["helpful_votes"] / votes_df["total_votes"] >= 0.5]
new_votes_df.head(30)
```

```python
paid_review = new_votes_df[new_votes_df["vine"]== "Y"]
paid_review
```

```python
unpaid_review = new_votes_df[new_votes_df["vine"]== "N"]
unpaid_review
```

```python
# Count total number of ratings for paid reviews
total_reviews = paid_review["review_id"].count()
total_reviews
```

```python
#Count number of ratings with 5 stars for paid reviews
five_stars_df = paid_review[paid_review["star_rating"] == 5.0]
five_stars = five_stars_df["star_rating"].count()
five_stars
```

```python
# Calculate percentage of 5-star reviews for paid reviews 
fivestart_percent = five_stars / total_reviews * 100
fivestart_percent
```

```python
# Create a function to find the total number of reviews, total number of 5 star reviews, 
# and percentage of 5 star reviews for each review type
def reviews(review_type):
    
    total = review_type["review_id"].count()
    
    five_stars_df = review_type[review_type["star_rating"] == 5.0]
    five_stars = five_stars_df["star_rating"].count()
    
    fivestar_percent = five_stars / total * 100
    
    return print (f"The total number of reviews is {total}.\nThe total number of 5 star reviews is {five_stars}.\nThe percentage of 5 star reviews is {fivestar_percent:.2f}%.")           
```

```python
reviews(paid_review)
```

```python
reviews(unpaid_review)
```
